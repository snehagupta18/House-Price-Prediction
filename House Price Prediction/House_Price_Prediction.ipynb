# House Price Prediction

## Project Goal

The project aims to predict the house prices using the California housing dataset. With the help of SVM, Linear Regression, Ridge, Lasso, and XGBRegressor, we endeavor to pinpoint the ideal model for predicting home values accurately.

##### Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn import metrics
```

##### Importing the California Housing Dataset

```python
from sklearn.datasets import fetch_california_housing
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to Pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
house_price_dataframe.head(5)
```

```python
# adding the target (price) column to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head(5)
```

```python
house_price_dataframe.shape
```

```python
# checking for any missing values present
house_price_dataframe.isnull().sum()
```

```python
# statistical analysis of the dataset
house_price_dataframe.describe()
```

##### Understanding the correlation between various features in the dataset

######
1. Positive Correlation

2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# using heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

##### Splitting the data and Target feature

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

##### Splitting the data into Training data and Test data

```python
#using 70% data for training purpose and 30% for testing purpose
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##### Model Training

### XGB Regressor

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

##### Evaluation

##### Prediction on training data

```python
# checking accuracy of prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#using R squared error and Mean Absolute error for accuracy of the model
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

##### Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

##### Prediction on Test Data

```python
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

### Linear regression

```python
model = LinearRegression()
model.fit(X_train, Y_train)
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

### SVM Regression Model

```python
from sklearn.svm import SVR
model = SVR(kernel='rbf')
model.fit(X_train, Y_train)
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

### Ridge Regression

```python
model = Ridge(alpha=1.0)
model.fit(X_train, Y_train)
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

### Lasso Regression

```python
model = Lasso(alpha=1.0)
model.fit(X_train, Y_train)
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
